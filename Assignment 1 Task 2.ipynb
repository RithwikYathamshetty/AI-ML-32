```python
#Assignment 1 Task 2
import matplotlib.pyplot as plt
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
# Scatter Plot
plt.scatter(age, height, color='blue')
plt.title('Scatter Plot of Age vs. Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.grid(True)
plt.show()
```

```python
# Bar Chart
plt.bar(people, weight, color='green')
plt.title('Bar Chart of Weight for Each Person')
plt.xlabel('People')
plt.ylabel('Weight')
plt.show()
```

```python
# Histogram
plt.hist(height, bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Height Distribution')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
healthcare_data = pd.DataFrame({'Name': people,'Age': age,'Height': height,'Weight': weight})

plt.figure(figsize=(8, 6))
plt.scatter(healthcare_data['Age'], healthcare_data['Weight'])
plt.title('Scatter Plot - Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
healthcare_data.plot(x='Name', y='Height', kind='bar', color='skyblue')
plt.title('Bar Chart - Height by Person')
plt.xlabel('Name')
plt.ylabel('Height')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(healthcare_data['Age'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Histogram - Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
